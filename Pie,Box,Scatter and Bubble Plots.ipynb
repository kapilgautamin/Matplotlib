```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("Canada.xlsx",
                  sheet_name="Canada by Citizenship",
                  skiprows = 20,
                  skipfooter = 2)
df
```

```python
df.shape
```

```python
df.drop(['AREA','REG','DEV','Type','Coverage'],axis=1,inplace=True)
df.rename(columns={'OdName':'Countries','AreaName':'Continents','RegName':'Region'},inplace=True)
df.head()
```

```python
df.set_index("Countries",inplace=True)
```

```python
years = list(range(1980,2014))
df['Total'] = df.sum(axis=1)
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df_cont = df.groupby('Continents',axis=0).sum()
df_cont
```

```python
df_cont['Total'].plot(kind="pie",
                      autopct='%1.1f%%',
                      startangle = 0,
                      shadow=True,
                     labels=None,
                     figsize=(15,6),
                     pctdistance=1.12)
plt.legend(df_cont.index.values,loc="upper left")
plt.axis('equal')
```

```python
df_2013 = df.groupby('Continents').sum()
df_2013[2013].plot(kind='pie',
                     figsize=(15,6),
                    autopct = '%1.1f%%',
                    pctdistance = 1.2,
                    startangle = 0,
                    labels=None)
plt.legend(df_2013.index.values,loc= "upper left")
plt.axis('equal')
```

Let's now try out the bubble plots

```python
dfT = df[years].T
dfT.index.name = 'year'
dfT.reset_index(inplace=True)
dfT.head()
```

```python
# normalize Brazil data
norm_brazil = (dfT['Brazil'] - dfT['Brazil'].min()) / (dfT['Brazil'].max() - dfT['Brazil'].min())

# normalize Argentina data
norm_argentina = (dfT['Argentina'] - dfT['Argentina'].min()) / (dfT['Argentina'].max() - dfT['Argentina'].min())
```

```python
# Brazil
ax0 = dfT.plot(kind='scatter',
                    x='year',
                    y='Brazil',
                    figsize=(14, 8),
                    alpha=0.5,                  # transparency
                    color='green',
                    s=norm_brazil * 2000 + 10,  # pass in weights 
                    xlim=(1975, 2015)
                   )

# Argentina
ax1 = dfT.plot(kind='scatter',
                    x='year',
                    y='Argentina',
                    alpha=0.5,
                    color="blue",
                    s=norm_argentina * 2000 + 10,
                    ax = ax0
                   )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```
